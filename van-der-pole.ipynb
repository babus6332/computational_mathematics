### Уравнение Ван-дер-Поля(решить уравнение и построить графики при параметре $a$)

#### Запишем уравнение в виде:


$
y_1^, = y_2 ;$

$ y_2^, = -a(y_2(y_1^2 - 1) + y_1)
$

Считаем, что параметр $a$ большой.
Для тестов рассматриваем $a = 1000, y_1 = 2, y_2 =0$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

a = 1000

def van_der_pol(x, t):
    nx0 = x[1]
    nx1 = -a * (x[0] ** 2.0 - 1.0) * x[1] - x[0]
    res = np.array([nx0, nx1])
    return res

start = 0.0
end = 100000.
step = 1000000

time = np.linspace(start, end, step)

plt.figure()
plt.grid()
# для наглядности строим графики с различными начальными точками (в задаче требует только голубой график)
xs = odeint(van_der_pol, [2., 0.], time)
plt.plot(xs[:,0], xs[:,1]) # голубой
xs = odeint(van_der_pol, [1.0, 2.0], time)
plt.plot(xs[:,0], xs[:,1]) #оранжевый
xs = odeint(van_der_pol, [4.0, 4.0], time)
plt.plot(xs[:,0], xs[:,1]) # зелёный
plt.gca().set_aspect('auto')  # можно выбрать equal
plt.xlabel(u'Координата y1')
plt.ylabel(u'Координата y2')
#plt.savefig('vanderpol_oscillator.png')
plt.show()
```
